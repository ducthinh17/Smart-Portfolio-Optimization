```python
import numpy as np
import pandas as pd
import cvxpy as cp

# 'bank_stocks' là DataFrame được khởi tạo ở trên
# Lọc ra chỉ các cột chứa giá đóng cửa (hoặc giá khác tùy theo nhu cầu)
# Ví dụ, nếu tên các cột giá đóng cửa là 'close_PLX', 'close_CRE', và 'close_VIB', 'close_SSI'
close_columns = [col for col in bank_stocks.columns if 'close' in col]
returns = bank_stocks[close_columns].pct_change().dropna()

# Tính toán ma trận covariance của lợi nhuận
cov_matrix = returns.cov()
expected_returns = returns.mean()

# Số lượng cổ phiếu
n_assets = len(returns.columns)

# Tạo biến tối ưu hóa
x = cp.Variable(n_assets)

# Mục tiêu: Tối đa hóa lợi nhuận kỳ vọng
objective = cp.Maximize(cp.sum(cp.multiply(expected_returns, x)))

# Ràng buộc: Tổng trọng số phải bằng 1, không có trọng số âm, rủi ro không vượt quá ngưỡng
risk_limit = 0.2  # Ngưỡng rủi ro, có thể điều chỉnh
constraints = [cp.sum(x) == 1, x >= 0, cp.quad_form(x, cov_matrix) <= risk_limit]

# Định nghĩa bài toán
problem = cp.Problem(objective, constraints)

# Giải bài toán
problem.solve()

# In kết quả
print("Trọng số tối ưu:", x.value)
```

```python
# Giả định optimal_weights là dữ liệu của bạn
optimal_weights = np.array([7.41456701e-09, 5.55464227e-08, 2.84218796e-08, 9.99999909e-01])
total_investment = 10000  # Số tiền bạn muốn đầu tư

# Lấy giá đóng cửa gần nhất cho mỗi cổ phiếu
last_PLX = PLX['close'].tail(1).values[0]
last_CRE = CRE['close'].tail(1).values[0]
last_VIB = VIB['close'].tail(1).values[0]
last_SSI = SSI['close'].tail(1).values[0]


latest_prices = np.array([last_PLX, last_CRE, last_VIB,last_SSI])

# Tính toán số tiền để đầu tư vào mỗi cổ phiếu
investment_per_stock = total_investment * optimal_weights

# Tính toán số lượng cổ phiếu cần mua cho mỗi loại
shares_to_buy = investment_per_stock / latest_prices

print("Số tiền đầu tư cho mỗi cổ phiếu:", investment_per_stock)
print("Số lượng cổ phiếu cần mua:", shares_to_buy)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'returns' is a DataFrame containing the daily returns of stocks
# And 'optimal_weights' is the optimized weights from your model

# Calculate the daily returns for VNINDEX
vn_index['market_returns'] = vn_index['close'].pct_change().dropna()

# Calculate daily portfolio returns
portfolio_returns = returns.dot(optimal_weights)

# Calculate annualized return and standard deviation (risk) for the portfolio
annualized_return = np.mean(portfolio_returns) * 252
annualized_std = np.std(portfolio_returns) * np.sqrt(252)

print("Annualized return of the portfolio:", annualized_return)
print("Annualized standard deviation (risk) of the portfolio:", annualized_std)

# Calculate the annualized return and standard deviation (risk) for the market
market_annualized_return = np.mean(vn_index['market_returns']) * 252
market_annualized_std = np.std(vn_index['market_returns']) * np.sqrt(252)

print("Annualized return of the market:", market_annualized_return)
print("Annualized standard deviation (risk) of the market:", market_annualized_std)

# Calculate Sharpe Ratio for both portfolio and market
# Assuming the risk-free rate is 0 for simplicity, replace '0' with the actual risk-free rate for more accuracy
risk_free_rate = 0
portfolio_sharpe_ratio = (annualized_return - risk_free_rate) / annualized_std
market_sharpe_ratio = (market_annualized_return - risk_free_rate) / market_annualized_std

print("Sharpe Ratio of the portfolio:", portfolio_sharpe_ratio)
print("Sharpe Ratio of the market:", market_sharpe_ratio)

# Visualize cumulative returns
cumulative_portfolio_returns = (1 + portfolio_returns).cumprod()
cumulative_market_returns = (1 + vn_index['market_returns']).cumprod()

plt.figure(figsize=(14, 7))
plt.plot(cumulative_portfolio_returns, label='Optimized Portfolio')
plt.xlabel('Time')
plt.ylabel('Cumulative Returns')
plt.title('Cumulative Returns of Portfolio vs Market')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np


# Lấy dữ liệu giá đóng cửa
closing_prices = bank_stocks.xs(key='close', level='Stock Info', axis=1)

# Tính lợi tức hàng ngày
returns = closing_prices.pct_change().dropna()

def calculate_portfolio_return(weights, returns):
    return np.dot(weights, returns.mean(axis=0))

def calculate_risk(weights, covariance_matrix):
    return np.sqrt(np.dot(weights.T, np.dot(covariance_matrix, weights)))

def gradient(weights, returns, covariance_matrix):
    average_returns = returns.mean(axis=0)
    portfolio_return = calculate_portfolio_return(weights, returns)
    risk = calculate_risk(weights, covariance_matrix)
    grad = (covariance_matrix @ weights) * risk - average_returns
    return grad

def sgd_portfolio_optimization(returns, initial_weights, learning_rate, iterations, covariance_matrix):
    weights = np.array(initial_weights)
    for _ in range(iterations):
        grad = gradient(weights, returns, covariance_matrix)
        weights -= learning_rate * grad
        weights = np.maximum(weights, 0)
        weights /= np.sum(weights)
    return weights

# Ma trận hiệp phương sai
covariance_matrix = returns.cov()

# Thực hiện tối ưu hóa danh mục đầu tư
initial_weights = [0.25, 0.25, 0.25, 0.25]  # Điều chỉnh trọng số ban đầu nếu cần
learning_rate = 0.01
iterations = 1000
optimal_weights = sgd_portfolio_optimization(returns, initial_weights, learning_rate, iterations, covariance_matrix)

print("Trọng số tối ưu cho danh mục đầu tư:", optimal_weights)
```

### Mô phỏng Monte Carlo

Monte Carlo:
1. Mục Đích: Phương pháp này sử dụng mô phỏng ngẫu nhiên để dự đoán các kết quả tương lai dựa trên một số giả định về biến động và lợi nhuận trung bình.
2. Ứng Dụng: Trong quản lý rủi ro tài chính, nó giúp đánh giá rủi ro và hiệu suất tiềm năng của danh mục đầu tư.
- Số lượng Mô phỏng (n_simulations):
Lý Do Chọn 10000: Số lượng mô phỏng là 10000 giúp cân bằng giữa độ chính xác và thời gian tính toán. Một số lượng lớn các mô phỏng đảm bảo rằng kết quả phân tích phản ánh một phạm vi rộng lớn của các kịch bản có thể xảy ra.
- Độ Chính Xác: Số lượng mô phỏng lớn giúp giảm thiểu "nhiễu" ngẫu nhiên và cung cấp một ước lượng chính xác hơn về kỳ vọng và rủi ro của danh mục.

```python
#Bứớc 4
import numpy as np

# Số lần mô phỏng
n_simulations = 10000
n_days = 252  # Số ngày giao dịch trong một năm

# Lưu trữ kết quả mô phỏng
portfolio_returns = np.empty(n_simulations)

# Thực hiện mô phỏng
for i in range(n_simulations):
    simulated_daily_returns = np.random.normal(loc=expected_return, scale=portfolio_risk, size=n_days)
    simulated_portfolio_return = np.prod(1 + simulated_daily_returns) - 1
    portfolio_returns[i] = simulated_portfolio_return

# Tính lợi nhuận trung bình và độ lệch chuẩn từ mô phỏng
mean_return = np.mean(portfolio_returns)
std_deviation = np.std(portfolio_returns)

# Tính VaR và CVaR
confidence_level = 0.05
VaR = np.percentile(portfolio_returns, 100 * confidence_level)
CVaR = np.mean(portfolio_returns[portfolio_returns <= VaR])

print(f"Trung bình lợi nhuận từ mô phỏng: {mean_return}")
print(f"Độ lệch chuẩn từ mô phỏng: {std_deviation}")
print(f"VaR ở mức tin cậy {100 * (1 - confidence_level)}%: {VaR}")
print(f"CVaR ở mức tin cậy {100 * (1 - confidence_level)}%: {CVaR}")
```

**Rủi ro VaR và CVaR:**
1. VaR (Value at Risk): Đo lường mức độ lỗ tối đa dự kiến ​​trong một khoảng thời gian nhất định, với một mức độ tin cậy nhất định (ví dụ, 95%).
2. CVaR (Conditional Value at Risk): Tính trung bình lỗ trong những trường hợp xấu nhất (tức là những trường hợp nằm trong 5% kịch bản tồi tệ nhất nếu VaR là 95%).

### Performance Visualization
